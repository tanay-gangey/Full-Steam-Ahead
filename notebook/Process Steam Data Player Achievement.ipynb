```python
import pandas as pd
import numpy as np

from tqdm import tqdm

tqdm.pandas()
```

```python
player_achievement = pd.read_csv("../steam-data/player_achievements_processed.csv", error_bad_lines=False, header=None)
```

```python
player_achievement.shape
```

```python
player_achievement.head()
```

```python
player_achievement.columns = ["steamid", "appid", "appname", "achievement"]
```

```python
achievement_percent = pd.read_csv("../steam-data/Achievement_Percentages.csv", error_bad_lines=False, header=None)
```

```python
achievement_percent.columns = ["unused", "appid", "achievement", "percent_achieved"]
```

```python
achievement_percent = achievement_percent.drop(["unused"], axis = 1)
```

```python
achievement_percent
```

```python
player_achievement.head()
```

```python
player_achievement = player_achievement.merge(achievement_percent, on=["appid", "achievement"], how="left").dropna()
```

```python
player_achievement.head()
```

```python
games = pd.read_csv("../steam-data/Games.csv", error_bad_lines=False, header=None)
```

```python
games.columns = ["idx", "appid", "total_playtime", "unused1", "unused2", "unused3", "steamid", "unused4"]
```

```python
games = games.drop(["idx",  "unused1", "unused2", "unused3", "unused4"], axis=1)
```

```python
games
```

```python
combined = games.merge(player_achievement, on=["appid", "steamid"])
```

```python
combined.to_csv("../steam-data/combined_achievement_data.csv")
```

```python
playtime_vs_achievement = combined.groupby(["appid", "appname","steamid","total_playtime"]).agg({"achievement": "count"}).reset_index()
```

```python
playtime_vs_achievement
```

```python
top_10_apps = playtime_vs_achievement['appid'].value_counts().nlargest(10)
```

```python
top_10_apps.index
```

```python
playtime_vs_achievement[playtime_vs_achievement['appid'].isin(top_10_apps.index)]
```

```python
top_10_apps
```

```python
lowest_percentage_achievement = combined[combined['percent_achieved']!=0]['percent_achieved'].nsmallest(10).reset_index()['percent_achieved']
```
