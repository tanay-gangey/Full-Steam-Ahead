```python
import numpy as np
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
```

```python
%cd ../steam-data/
```

```python
df = pd.read_csv("NewFriends.csv")
df = df[df['year'] > 2018]
df = df.groupby(['year', 'month']).count().reset_index()
df['year-month'] =df['year'].astype(str)+ "-" + df['month'].astype(str)
df = df[df['year-month'] != '2021-12']
display(df)
df.plot.line(x='year-month', y='user_id', xlabel='Year-Month', ylabel='Friends Count', title='Steam Friendship Trends')
```
